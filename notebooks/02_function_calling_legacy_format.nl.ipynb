## 🇳🇱 Historische Context
Het legacy formaat was de eerste manier om externe functies aan te roepen met OpenAI modellen. Het gebruikte:
- `functions` parameter (nu vervangen door `tools`)
- `function_call` parameter (nu vervangen door `tool_choice`)
- Enkele functie-aanroepen per keer (nu meerdere mogelijk)

## 🎯 Leerdoelen
- Begrijpen waarom het formaat is geëvolueerd
- Herkennen van legacy code in bestaande projecten
- Migreren naar het moderne formaat

````python
import os
import json
from openai import AzureOpenAI
from dotenv import load_dotenv
load_dotenv()

# Azure OpenAI Configuratie
client = AzureOpenAI(
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    api_version="2024-05-01-preview"
)

deployment = os.getenv("AZURE_OPENAI_GPT4O_DEPLOYMENT_NAME")

print("✅ Azure OpenAI client geïnitialiseerd")
print("⚠️ Dit notebook demonstreert het LEGACY formaat - gebruik dit niet in productie!")
````

## 🔧 Legacy vs Modern Formaat Vergelijking

### Legacy Formaat (Verouderd):
```python
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": "Hoe is het weer in Amsterdam?"}],
    functions=[  # ❌ Verouderd
        {
            "name": "get_weather",
            "parameters": {...}
        }
    ],
    function_call="auto"  # ❌ Verouderd
)
```

### Modern Formaat (Aanbevolen):
```python
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": "Hoe is het weer in Amsterdam?"}],
    tools=[  # ✅ Modern
        {
            "type": "function",
            "function": {
                "name": "get_weather",
                "parameters": {...}
            }
        }
    ],
    tool_choice="auto"  # ✅ Modern
)
```

````python
# Legacy functie definitie (VEROUDERD - alleen ter demonstratie)
legacy_functions = [
    {
        "name": "get_dutch_weather",
        "description": "Verkrijg het weer voor een Nederlandse stad",
        "parameters": {
            "type": "object",
            "properties": {
                "city": {
                    "type": "string",
                    "description": "De stad, bijv. Amsterdam, Rotterdam"
                }
            },
            "required": ["city"]
        }
    }
]

print("⚠️ Legacy functie definities:")
print(json.dumps(legacy_functions, indent=2, ensure_ascii=False))
print("\n❌ Dit formaat is verouderd en wordt mogelijk niet meer ondersteund!")
````

## 🔄 Migratie naar Modern Formaat

Hier is hoe je legacy code kunt migreren naar het moderne formaat:

````python
def convert_legacy_to_modern(legacy_functions):
    """Converteer legacy function format naar modern tools format"""
    
    modern_tools = []
    
    for func in legacy_functions:
        modern_tool = {
            "type": "function",
            "function": {
                "name": func["name"],
                "description": func.get("description", ""),
                "parameters": func["parameters"]
            }
        }
        modern_tools.append(modern_tool)
    
    return modern_tools

# Converteer legacy naar modern
modern_tools = convert_legacy_to_modern(legacy_functions)

print("✅ Geconverteerd naar modern formaat:")
print(json.dumps(modern_tools, indent=2, ensure_ascii=False))
````

## 📊 Vergelijking: Legacy vs Modern

| Aspect | Legacy Formaat | Modern Formaat |
|--------|---------------|----------------|
| Parameter naam | `functions` | `tools` |
| Selectie param | `function_call` | `tool_choice` |
| Meerdere aanroepen | ❌ Nee | ✅ Ja |
| Type specificatie | Impliciet | Expliciet (`type: "function"`) |
| Response formaat | `function_call` | `tool_calls` array |
| Ondersteuning | Afgeschreven | Actief onderhouden |
