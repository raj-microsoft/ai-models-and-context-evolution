## Setup: Initialize Azure OpenAI Client

Let's set up our Azure OpenAI client to enable Q&A functionality.

```python
import os
import json
from openai import AzureOpenAI
from dotenv import load_dotenv
load_dotenv()

# Azure OpenAI Configuration
endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
deployment = os.getenv("AZURE_OPENAI_GPT4O_DEPLOYMENT_NAME")
subscription_key = os.getenv("AZURE_OPENAI_API_KEY")
api_version = os.getenv("AZURE_OPENAI_GPT4O_API_VERSION")

# Initialize Azure OpenAI client
client = AzureOpenAI(
    azure_endpoint=endpoint,
    api_key=subscription_key,
    api_version=api_version,
)

print("✅ Azure OpenAI client initialized successfully!")
print(f"📍 Endpoint: {endpoint}")
print(f"🤖 Deployment: {deployment}")
```

## Load GitHub Universe 2025 Data

Load the announcement data from our JSON file to provide context for Q&A.

```python
# Load GitHub Universe 2025 announcements
with open('../assets/github_universe_2025_announcements.json', 'r') as f:
    github_universe_data = json.load(f)

print("📚 Loaded GitHub Universe 2025 Data")
print(f"Event: {github_universe_data['event']}")
print(f"\nMajor Announcements: {len(github_universe_data['major_announcements'])}")
print("\nAnnouncement Titles:")
for announcement in github_universe_data['major_announcements']:
    print(f"  - {announcement['title']} ({announcement['status']})")
```

## Helper Function: Ask Questions About GitHub Universe 2025

This function uses the loaded data as context to answer questions about GitHub Universe 2025.

```python
def ask_about_github_universe(question):
    """
    Ask a question about GitHub Universe 2025 announcements.
    Uses the loaded data as context for accurate answers.
    """
    # Create system prompt with GitHub Universe data as context
    system_prompt = f"""
You are a knowledgeable assistant about GitHub Universe 2025. 
Use the following official information to answer questions accurately.

GitHub Universe 2025 Data:
{json.dumps(github_universe_data, indent=2)}

When answering:
- Be specific and reference exact features or statistics from the data
- Cite the status (Generally Available, Public Preview, etc.) when relevant
- Mention official resources when appropriate
- If asked about something not in the data, acknowledge that limitation
"""
    
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": question}
    ]
    
    response = client.chat.completions.create(
        model=deployment,
        messages=messages,
        max_tokens=1000,
        temperature=0.7
    )
    
    return response.choices[0].message.content

print("✅ Q&A function ready! Use ask_about_github_universe('your question') to get answers.")
```
